```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
loan_detail_df = pd.read_csv('data/sending/loan_details_cleaned.csv')
metadata_df = pd.read_csv('data/sending/metadata.csv')
geographic_df = pd.read_csv('data/sending/geographic.csv')
financial_df = pd.read_csv('data/sending/financial_cleaned.csv')
demographic_df = pd.read_csv('data/sending/demographic_cleaned.csv')
credit_history_df = pd.read_csv('data/sending/credit_history_cleaned.csv')
```

```python
merged_meta_demo_df = metadata_df.merge(demographic_df, on='customer_id', how='left')
```

```python
loan_meta_demo_df = merged_meta_demo_df.merge(loan_detail_df, on='customer_id', how='left')
```

```python
credit_loan_meta_demo_df = loan_meta_demo_df.merge(credit_history_df, how='left', on='customer_id')
```

```python
merged_df = credit_loan_meta_demo_df.merge(financial_df, how='left', on='customer_id')
```

```python
merged_df.to_csv('data/merged_final.csv', index=False)
```
