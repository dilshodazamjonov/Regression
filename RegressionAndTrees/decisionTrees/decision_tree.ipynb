```python
import pandas as pd

data = pd.read_csv('data/pokemon.csv')
data = data.rename(columns={"Type 1": "Type"})
data = data.query("Type.isin(('Electric', 'Grass'))")

X = data[["HP", "Attack", "Defense", "Speed"]]
y = (data["Type"] == "Electric")

data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt


tree = DecisionTreeClassifier(max_depth=3, criterion = 'entropy').fit(X, y)
plt.figure(figsize=(12,8))
plot_tree(tree, filled=True, feature_names=X.columns, class_names=True)
plt.show()

```

```python
# accuracy score
from sklearn.metrics import accuracy_score

accuracy_score(y, tree.predict(X))
```
