```python
from sklearn.datasets import make_swiss_roll, make_s_curve, make_circles, make_moons
import matplotlib.pyplot as plt

# 1. Swiss Roll
X, _ = make_swiss_roll(n_samples=2000, noise=0.05)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], s=5)
plt.title("Swiss Roll Data")
plt.show()
```

```python
# 2. S-curve
X, _ = make_s_curve(n_samples=2000, noise=0.05)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], s=5)
plt.title("S-Curve")
plt.show()
```

```python
# 3. Concentric circles
X, y = make_circles(n_samples=500, factor=0.4, noise=0.05)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Concentric Circles")
plt.show()
```

```python

# 4. Two moons
X, y = make_moons(n_samples=500, noise=0.05)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Two Moons")
plt.show()
```
